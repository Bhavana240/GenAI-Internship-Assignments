```python
# ==========================================================
# NLP Assignment 3: Chatbot using Hugging Face Transformers
# Model: DialoGPT (Microsoft)
# Author: Bhavana S
# Description: Console-based AI chatbot with conversation memory
# ==========================================================


# ===============================
# STEP 1: INSTALL REQUIRED LIBRARIES
# ===============================
!pip install transformers torch


# ===============================
# STEP 2: IMPORT LIBRARIES
# ===============================
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
import time


# ===============================
# STEP 3: LOAD PRE-TRAINED MODEL
# ===============================
def load_model():
    """
    Load tokenizer and model from Hugging Face
    """
    print("🔄 Loading DialoGPT model... Please wait...\n")

    tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
    model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

    print("✅ Model loaded successfully!\n")
    return tokenizer, model


# ===============================
# STEP 4: DISPLAY INTRO MESSAGE
# ===============================
def display_intro():
    """
    Show chatbot introduction
    """
    print("="*60)
    print("🤖 AI CHATBOT USING TRANSFORMERS")
    print("="*60)
    print("Chatbot: Hello! I am your AI assistant.")
    print("Chatbot: Ask me anything 😊")
    print("Type 'exit' or 'quit' to stop the chatbot.\n")


# ===============================
# STEP 5: GENERATE RESPONSE FUNCTION
# ===============================
def generate_response(user_input, tokenizer, model, chat_history_ids):
    """
    Generate chatbot response using transformer model
    """

    # Encode input with end-of-string token
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    # Append conversation history
    if chat_history_ids is not None:
        bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
    else:
        bot_input_ids = new_input_ids

    # Generate response from model
    chat_history_ids = model.generate(
        bot_input_ids,
        max_length=1000,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        top_k=100,
        top_p=0.75,
        temperature=0.85,
        no_repeat_ngram_size=3
    )

    # Decode only the new response
    response = tokenizer.decode(
        chat_history_ids[:, bot_input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )

    return response, chat_history_ids


# ===============================
# STEP 6: TYPING EFFECT (FOR BETTER UX)
# ===============================
def typing_effect(text):
    """
    Simulate chatbot typing effect
    """
    for char in text:
        print(char, end='', flush=True)
        time.sleep(0.02)
    print()


# ===============================
# STEP 7: MAIN CHAT FUNCTION
# ===============================
def start_chat():
    """
    Main function to run chatbot
    """

    # Load model
    tokenizer, model = load_model()

    # Show intro
    display_intro()

    # Initialize chat history
    chat_history_ids = None

    # Store full conversation (optional feature)
    conversation_log = []

    # Start conversation loop
    while True:

        # Take user input
        user_input = input("You: ")

        # Exit condition
        if user_input.lower() in ['exit', 'quit']:
            print("\nChatbot: Goodbye! Have a great day 😊")
            break

        # Store user input
        conversation_log.append("User: " + user_input)

        # Generate chatbot response
        response, chat_history_ids = generate_response(
            user_input, tokenizer, model, chat_history_ids
        )

        # Store chatbot response
        conversation_log.append("Bot: " + response)

        # Display response with typing effect
        print("Chatbot: ", end="")
        typing_effect(response)


# ===============================
# STEP 8: RUN THE CHATBOT
# ===============================
if __name__ == "__main__":
    start_chat()


# ===============================
# STEP 9 (OPTIONAL): PRINT FULL CHAT HISTORY
# ===============================
def show_conversation_log(conversation_log):
    """
    Display entire conversation history
    """
    print("\n📜 FULL CONVERSATION HISTORY:\n")
    for line in conversation_log:
        print(line)
```
