```python
# ============================================================
# INSTALL REQUIRED LIBRARY
# ============================================================
!pip install langchain

# ============================================================
# IMPORTS
# ============================================================
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

# ============================================================
# TASK 1: Replace Hardcoded Prompt with PromptTemplate
# ============================================================

# Creating a reusable PromptTemplate (NO f-strings used)
template_task1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Function to use the template
def explain_topic(topic):
    return template_task1.format(topic=topic)

# Test
print("TASK 1 OUTPUT:")
print(explain_topic("Artificial Intelligence"))
print("\n")


# ============================================================
# TASK 2: Multi-Input Prompt System
# ============================================================

template_task2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def multi_input_prompt(topic, audience, tone):
    return template_task2.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test cases
print("TASK 2 OUTPUT:")
print(multi_input_prompt("AI", "beginners", "friendly"))
print(multi_input_prompt("Python", "kids", "fun"))
print(multi_input_prompt("Deep Learning", "engineers", "technical"))
print("\n")


# ============================================================
# TASK 3: Prompt Variations Engine
# ============================================================

# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

def run_variations(topic):
    print("Teaching Style:")
    print(teaching_template.format(topic=topic))

    print("\nInterview Style:")
    print(interview_template.format(topic=topic))

    print("\nStorytelling Style:")
    print(story_template.format(topic=topic))

print("TASK 3 OUTPUT:")
run_variations("Machine Learning")
print("\n")


# ============================================================
# TASK 4: ChatPromptTemplate System (ROLE-BASED)
# ============================================================

def chat_prompt(role, topic):
    # Define system behavior based on role
    system_template = "You are a {role} who explains topics clearly."
    human_template = "Explain {topic}"

    system_message = SystemMessagePromptTemplate.from_template(system_template)
    human_message = HumanMessagePromptTemplate.from_template(human_template)

    chat_template = ChatPromptTemplate.from_messages([system_message, human_message])

    # Format messages
    return chat_template.format_messages(role=role, topic=topic)

print("TASK 4 OUTPUT:")
messages = chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
print("\n")


# ============================================================
# TASK 5: Input Validation Layer
# ============================================================

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Check audience
    if audience not in valid_audience:
        print("Invalid audience! Setting default: beginner")
        audience = "beginner"

    # Check tone
    if tone not in valid_tone:
        print("Invalid tone! Setting default: formal")
        tone = "formal"

    return audience, tone

print("TASK 5 OUTPUT:")
print(validate_inputs("kids", "cool"))   # invalid example
print("\n")


# ============================================================
# TASK 6: Prompt Generator App
# ============================================================

# Single reusable template
generator_template = PromptTemplate(
    input_variables=["topic", "audience", "tone", "style"],
    template="Explain {topic} for {audience} in a {tone} {style} style"
)

def generate_prompt(topic, audience, tone, style):
    # Validate inputs first
    audience, tone = validate_inputs(audience, tone)

    # Generate prompt
    return generator_template.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
    )

print("TASK 6 OUTPUT:")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print("\n")


# ============================================================
# TASK 7: Template Reusability Test
# ============================================================

print("TASK 7 OUTPUT:")

topics = ["AI", "Python", "Data Science", "Cyber Security", "IoT"]

for t in topics:
    result = generator_template.format(
        topic=t,
        audience="beginner",
        tone="casual",
        style="teaching"
    )
    print(result)
```
